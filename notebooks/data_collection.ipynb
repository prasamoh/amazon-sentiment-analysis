```python
#exploring data only for Jan month

date_ranges = [
    ("2026-01-01", "2026-01-06"),
    ("2026-01-07", "2026-01-12"),
    ("2026-01-13", "2026-01-18"),
    ("2026-01-19", "2026-01-24"),
    ("2026-01-25", "2026-01-31"),
]
```

```python
#download data using ApifyClient for all tweets contain the keyword Amazon

from apify_client import ApifyClient
import pandas as pd

APIFY_TOKEN = "APIFY_TOKEN" # token removed
client = ApifyClient(APIFY_TOKEN)

ACTOR_ID = "apidojo/tweet-scraper"

all_items = []

for i, (start, end) in enumerate(date_ranges):
    print(f"\nBatch {i+1}: {start} → {end}")

    query = f'Amazon lang:en since:{start} until:{end} -filter:replies'

    run_input = {
        "searchTerms": [query],
        "sort": "Latest",
        "tweetsDesired": 800   # SAFE: ~500–800 per batch
    }

    run = client.actor(ACTOR_ID).call(run_input=run_input)

    dataset_id = run["defaultDatasetId"]
    items = list(client.dataset(dataset_id).iterate_items())

    print(f"Collected: {len(items)} tweets")

    all_items.extend(items)

    # 🔴 Safety stop if needed
    if len(all_items) > 3000:
        print("Stopping early to control cost")
        break
```

```python
#above jan is successful hence extended the search to Feb and March

feb_mar_ranges = [
    # February
    ("2026-02-01", "2026-02-06"),
    ("2026-02-07", "2026-02-12"),
    ("2026-02-13", "2026-02-18"),
    ("2026-02-19", "2026-02-24"),
    ("2026-02-25", "2026-02-28"),

    # March
    ("2026-03-01", "2026-03-06"),
    ("2026-03-07", "2026-03-12"),
    ("2026-03-13", "2026-03-18"),
    ("2026-03-19", "2026-03-24"),
    ("2026-03-25", "2026-03-31"),
]
```

```python
#download data using ApifyClient for all tweets contain the keyword Amazon

for i, (start, end) in enumerate(feb_mar_ranges):
    print(f"\nBatch {i+1}: {start} → {end}")

    query = f'(Amazon OR @amazon OR #amazon) lang:en since:{start} until:{end} -filter:replies'

    run_input = {
        "searchTerms": [query],
        "sort": "Latest",
        "tweetsDesired": 250,
        "disableMaximization": True
    }

    run = client.actor(ACTOR_ID).call(run_input=run_input)
    dataset_id = run["defaultDatasetId"]

    items = list(client.dataset(dataset_id).iterate_items())

    print(f"Collected: {len(items)} tweets")

    all_items.extend(items)

    # 🔴 Optional safety break
    if len(all_items) > 10000:
        print("Reached 10K target, stopping early")
        break

print("\nTotal raw collected:", len(all_items))
```

```python
#download data as JSON format for futher processing.

import json

with open("amazon_raw_data.json", "w") as f:
    json.dump(all_items, f)

print("Raw JSON saved")
```
